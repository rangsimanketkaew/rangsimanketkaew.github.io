```
#xrange v.s. range
# ****For Python 2****
#-----------------

import time

start = time.time()
# slower 
x = [i for i in range(0,10,2)] 
print(x)
print("Time elapsed for range is:", time.time() - start)

start = time.time()
# faster 
x = [i for i in xrange(0,10,2)] 
print(x)
print("Time elapsed for xrange is:", time.time() - start)

```
